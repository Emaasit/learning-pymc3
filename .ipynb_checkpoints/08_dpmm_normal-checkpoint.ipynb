# Dirichlet Process Mixture Models

```python
%matplotlib inline
import numpy as np
import pandas as pd
import theano
import theano.tensor as tt
import pymc3 as pm
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_context('notebook')
np.random.seed(12345)
rc = {'xtick.labelsize': 20, 'ytick.labelsize': 20, 'axes.labelsize': 20, 'font.size': 20, 
      'legend.fontsize': 12.0, 'axes.titlesize': 10, "figure.figsize": [14, 6]}
sns.set(rc = rc)
sns.set_style("whitegrid")
```

## Step 1: Prepare the data

```python
import statsmodels as sm
old_faithful_df = sm.datasets.get_rdataset('faithful', cache=True).data[['waiting']]
```

```python
old_faithful_df.head()
```

```python
old_faithful_df['std_waiting'] = (old_faithful_df.waiting - old_faithful_df.waiting.mean()) / old_faithful_df.waiting.std()
old_faithful_df.head()
```

```python
import plotly as py
fig, ax = plt.subplots()
ax = sns.distplot(bins = 30, kde = False, a = old_faithful_df.std_waiting)
ax.set_xlabel('Standardized waiting time between eruptions')
ax.set_ylabel('Number of eruptions');
#py.offline.iplot_mpl(fig)
plt.savefig('figpath1.svg', dpi = 400, bbox_inches = "tight")
```

```python
N = old_faithful_df.shape[0]
K = 30
```

## Step 2: Build the model

\begin{split}\begin{align*}
    \alpha
        & \sim \textrm{Gamma}(1, 1) \\
    \beta_1, \ldots, \beta_K
        & \sim \textrm{Beta}(1, \alpha) \\
    w_i
        & = \beta_i \prod_{j = i - 1}^i (1 - \beta_j) \\
    \\
    \lambda_1, \ldots, \lambda_K
        & \sim U(0, 5) \\
    \tau_1, \ldots, \tau_K
        & \sim \textrm{Gamma}(1, 1) \\
    \mu_i\ |\ \lambda_i, \tau_i
        & \sim N\left(0, (\lambda_i \tau_i)^{-1}\right) \\
    \\
    x\ |\ w_i, \lambda_i, \tau_i, \mu_i
        & \sim \sum_{i = 1}^K w_i\ N(\mu_i, (\lambda_i \tau_i)^{-1})
\end{align*}\end{split}

```python
def stick_breaking(beta):
    portion_remaining = tt.concatenate([[1], tt.extra_ops.cumprod(1 - beta)[:-1]])

    return beta * portion_remaining
```

```python
with pm.Model() as dpmm_model:
    alpha = pm.Gamma('alpha', 1., 1.)
    beta = pm.Beta('beta', 1., alpha, shape=K)
    w = pm.Deterministic('w', stick_breaking(beta))

    tau = pm.Gamma('tau', 1., 1., shape=K)
    lambda_ = pm.Uniform('lambda', 0, 5, shape=K)
    mu = pm.Normal('mu', 0, tau=lambda_ * tau, shape=K)
    obs = pm.NormalMixture('obs', w, mu, tau=lambda_ * tau,
                           observed=old_faithful_df.std_waiting.values)
```

## Step 3: Sample from the posterior

```python
with dpmm_model:
    #step = pm.NUTS(target_accept = 0.999, adapt_step_size = 0.001, max_treedepth = 15)
    posterior = pm.sample(draws = 500, njobs = 2)
```

## Step 4: Diagnose convergence of MCMC chains

```python
pm.traceplot(posterior)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

plot_w = np.arange(K) + 1

ax.bar(plot_w - 0.5, posterior['w'].mean(axis=0), width=1., lw=0);

ax.set_xlim(0.5, K);
ax.set_xlabel('Component');

ax.set_ylabel('Posterior expected mixture weight');
```

```python
x_plot = np.linspace(-3, 3, 200)
```

```python
import scipy as sp
post_pdf_contribs = sp.stats.norm.pdf(np.atleast_3d(x_plot),
                                      posterior['mu'][:, np.newaxis, :],
                                      1. / np.sqrt(posterior['lambda'] * posterior['tau'])[:, np.newaxis, :])
post_pdfs = (posterior['w'][:, np.newaxis, :] * post_pdf_contribs).sum(axis=-1)

post_pdf_low, post_pdf_high = np.percentile(post_pdfs, [2.5, 97.5], axis=0)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

n_bins = 20
ax.hist(old_faithful_df.std_waiting.values, bins=n_bins, normed=True,
        color="blue", lw=0, alpha=0.5);

ax.fill_between(x_plot, post_pdf_low, post_pdf_high,
                color='gray', alpha=0.45);
ax.plot(x_plot, post_pdfs[0],
        c='gray', label='Posterior sample densities');
ax.plot(x_plot, post_pdfs[::100].T, c='gray');
ax.plot(x_plot, post_pdfs.mean(axis=0),
        c='k', label='Posterior expected density');

ax.set_xlabel('Standardized waiting time between eruptions');

ax.set_yticklabels([]);
ax.set_ylabel('Density');

ax.legend(loc=2);
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

n_bins = 20
ax.hist(old_faithful_df.std_waiting.values, bins=n_bins, normed=True,
        color="blue", lw=0, alpha=0.5);

ax.plot(x_plot, post_pdfs.mean(axis=0),
        c='k', label='Posterior expected density');
ax.plot(x_plot, (posterior['w'][:, np.newaxis, :] * post_pdf_contribs).mean(axis=0)[:, 0],
        '--', c='k', label='Posterior expected mixture\ncomponents\n(weighted)');
ax.plot(x_plot, (posterior['w'][:, np.newaxis, :] * post_pdf_contribs).mean(axis=0),
        '--', c='k');

ax.set_xlabel('Standardized waiting time between eruptions');

ax.set_yticklabels([]);
ax.set_ylabel('Density');

ax.legend(loc=2);
```
